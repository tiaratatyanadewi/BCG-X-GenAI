```python
import pandas as pd
```

```python
df = pd.read_csv('final_data.csv')
```

```python
df
```

```python
def generate_qa_list():
    questions = []
    answers = []

    for i in range(len(df["Company"])):
        questions.append(f"What is the Total Revenue of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"${df['Total Revenue'][i]:,.0f}")

        questions.append(f"What is the Net Income of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"${df['Net Income'][i]:,.0f}")

        questions.append(f"What are the Total Assets of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"${df['Total Assets'][i]:,.0f}")

        questions.append(f"What are the Total Liabilities of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"${df['Total Liabilities'][i]:,.0f}")

        questions.append(f"What is the Cash Flow of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"${df['Cash Flow'][i]:,.0f}")

        questions.append(f"What is the Revenue Growth (%) of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"{df['Revenue Growth (%)'][i]:,.6f}%")

        questions.append(f"What is the Net Income Growth (%) of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"{df['Net Income Growth (%)'][i]:,.6f}%")

        questions.append(f"What is the Liabilities to Assets ratio of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"{df['Liabilities_to_Assets'][i]:,.6f}")

        questions.append(f"What is the Net Income to Revenue ratio of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"{df['Net_Income_to_Revenue'][i]:,.6f}")

        questions.append(f"What is the Revenue Growth Rate of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"{df['Revenue_Growth_Rate'][i]:,.6f}")

        questions.append(f"What is the Net Income Growth Rate of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"{df['Net_Income_Growth_Rate'][i]:,.6f}")

        questions.append(f"What is the Revenue Rolling Mean of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"${df['Revenue_Rolling_Mean'][i]:,.0f}" if df['Revenue_Rolling_Mean'][i] is not None else "0")

        questions.append(f"What is the Revenue Rolling Std of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"${df['Revenue_Rolling_Std'][i]:,.0f}" if df['Revenue_Rolling_Std'][i] is not None else "0")

        questions.append(f"What is the Net Income Rolling Mean of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"${df['Net_Income_Rolling_Mean'][i]:,.0f}" if df['Net_Income_Rolling_Mean'][i] is not None else "0")

        questions.append(f"What is the Net Income Rolling Std of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"${df['Net_Income_Rolling_Std'][i]:,.0f}" if df['Net_Income_Rolling_Std'][i] is not None else "0")

        questions.append(f"What is the Revenue Lag 1 of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"${df['Revenue_Lag_1'][i]:,.0f}" if df['Revenue_Lag_1'][i] is not None else "0")

        questions.append(f"What is the Net Income Lag 1 of {df['Company'][i]} in {df['Year'][i]}?")
        answers.append(f"${df['Net_Income_Lag_1'][i]:,.0f}" if df['Net_Income_Lag_1'][i] is not None else "0")

    qa_df = pd.DataFrame({"Question": questions, "Answer": answers})

    return qa_df
```

```python
qa_df = generate_qa_list()
qa_df.info()
```

```python
qa_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
qa_df.isnull().sum()
```

```python
vec = TfidfVectorizer()
vec.fit(np.concatenate((qa_df['Question'], qa_df['Answer'])))
```

```python
feature_col = vec.get_feature_names_out()
len(feature_col)
```

```python
df_vectors = vec.transform(qa_df['Question'])
```

```python
df_vectors
```

```python
print("Hello, welcome to BCG X Financial Chatbot Demo")
print("Ask me anything you want")

while True:
    input_question = input()

    if input_question.lower() == 'stop':
        break

    input_question_vec = vec.transform([input_question])
    similarity = cosine_similarity(input_question_vec, df_vectors)
    closest_ans = np.argmax(similarity, axis=1)

    # Check if the similarity is below a certain threshold
    max_similarity = similarity[0][closest_ans[0]]
    if max_similarity < 0.1:  # You can adjust this threshold based on your needs
        response = "Sorry, we found no data related to that question."
    else:
        response = qa_df['Answer'].iloc[closest_ans].values[0]

    print(f"Response from chat bot is: {response}")
```
